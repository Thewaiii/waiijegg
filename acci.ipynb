#### จัดกลุ่มแต่ละเวลาให้แบ่งเป็นช่วงเวลา

```python
import datetime
def time_in_range(start, end, x):
    """Return true if x is in the range [start, end]"""
    if start <= end:
        return start <= x <= end
    else:
        return start <= x or x <= end
```

```python
# Filter out rows with NaN in 'expw_step' as this column is used for several new features
filtered_data = use_data.dropna(subset=['expw_step']).reset_index(drop=True)

# Recreate lists from the filtered data
cov_time_sti_filtered = filtered_data['accident_date'].apply(lambda x: x.strftime("%Y-%m-%d %H:%M:%S"))

day_filtered = []
mon_filtered = []
year_filtered = []
for i in range(len(cov_time_sti_filtered)):
  L = cov_time_sti_filtered[i].split()[0].split('-')
  year_filtered.append(L[0])
  mon_filtered.append(L[1])
  day_filtered.append(L[2])

time_lable_filtered = []
#เช้า
start_0 = datetime.time(6, 0, 0)
end_0 = datetime.time(11, 59, 59)
#บ่าย
start_1 = datetime.time(12, 0, 0)
end_1 = datetime.time(17, 59, 59)
#ดึก
start_2 = datetime.time(18, 0, 0)
end_2 = datetime.time(23, 59, 59)
#หลังเที่ยงคืน
start_3 = datetime.time(0, 0, 0)
end_3 = datetime.time(5, 59, 59)
for i in range(len(filtered_data['accident_time'])):
  if time_in_range(start_0, end_0, filtered_data['accident_time'][i]) == True:
    time_lable_filtered.append('0')
  elif time_in_range(start_1, end_1, filtered_data['accident_time'][i]) == True:
    time_lable_filtered.append('1')
  elif time_in_range(start_2, end_2, filtered_data['accident_time'][i]) == True:
    time_lable_filtered.append('2')
  elif time_in_range(start_3, end_3, filtered_data['accident_time'][i]) == True:
    time_lable_filtered.append('3')


name_road_filtered=[]
for i in range(len(filtered_data['expw_step'])):
  if filtered_data['expw_step'][i] == 'ศรีรัช':
    name_road_filtered.append('0')
  elif filtered_data['expw_step'][i] == 'ฉลองรัช':
    name_road_filtered.append('1')
  elif filtered_data['expw_step'][i] == 'บูรพาวิถี':
    name_road_filtered.append('2')
  elif filtered_data['expw_step'][i] == 'อุดรรัถยา':
    name_road_filtered.append('3')
  elif filtered_data['expw_step'][i] == 'เฉลิมมหานคร':
    name_road_filtered.append('4')
  elif filtered_data['expw_step'][i] == 'บางพลี-สุขสวัสดิ์':
    name_road_filtered.append('5')
  elif filtered_data['expw_step'][i] == 'ทางหลวงพิเศษหมายเลข 37':
    name_road_filtered.append('6')
  elif filtered_data['expw_step'][i] == 'ศรีรัช-วงแหวนรอบนอก':
    name_road_filtered.append('7')
  elif filtered_data['expw_step'][i] == 'S1':
    name_road_filtered.append('8')


weather_road_filtered=[]
for i in range(len(filtered_data['weather_state'])):
  if filtered_data['weather_state'][i] == 'ปกติ':
    weather_road_filtered.append('0')
  elif (filtered_data['weather_state'][i] == 'ฝนตก') | (filtered_data['weather_state'][i] == 'ท้องฟ้ามืดครึ้ม') |(filtered_data['weather_state'][i] == 'ฝนตกลมแรง'):
    weather_road_filtered.append('1')
  elif filtered_data['weather_state'][i] == 'กลางคืน':
    weather_road_filtered.append('2')
  elif (filtered_data['weather_state'][i] == 'ถนนเปียกรื่น') |(filtered_data['weather_state'][i] == 'ถนนเปียกลื่น')|(filtered_data['weather_state'][i] == 'ถนนเปียก') :
    weather_road_filtered.append('3')


injur_filtered =[]
for i in range(len(filtered_data['injur_man'])):
  if (filtered_data['injur_man'][i] == 0 ) and (filtered_data['injur_femel'][i] == 0):
    injur_filtered.append('0')
  elif (filtered_data['injur_man'][i] > 0) or (filtered_data['injur_femel'][i] > 0):
    injur_filtered.append('1')


dead_filtered =[]
for i in range(len(filtered_data['dead_man'])):
  if (filtered_data['dead_man'][i] == 0 ) and (filtered_data['dead_femel'][i] == 0):
    dead_filtered.append('0')
  elif (filtered_data['dead_man'][i] > 0) or (filtered_data['dead_femel'][i] > 0):
    dead_filtered.append('1')


value_of_acdi_filtered =[]
for i in range(len(filtered_data['dead_man'])):
  if (filtered_data['dead_man'][i]+filtered_data['dead_femel'][i] == 0 ) and (filtered_data['injur_man'][i]+filtered_data['injur_femel'][i]==0):
    value_of_acdi_filtered.append('0')
  elif (filtered_data['dead_man'][i]+filtered_data['dead_femel'][i] > 0 ) or (filtered_data['injur_man'][i]+filtered_data['injur_femel'][i]>0):
    value_of_acdi_filtered.append('1')

# Apply cause grouping to the filtered data
new1_filtered = filtered_data['cause'].str.strip().replace(cause_group)


# สร้าง DataFrame ใหม่จาก lists ที่กรองแล้ว
new_df = pd.DataFrame({
    'year': year_filtered,
    'month': mon_filtered,
    'day': day_filtered,
    'time_label': time_lable_filtered,
    'road_name': name_road_filtered,
    'weather_state': weather_road_filtered,
    'injured': injur_filtered,
    'dead': dead_filtered,
    'accident_severity': value_of_acdi_filtered,
    'cause': new1_filtered
})

print("\nDataFrame ใหม่ (new_df) หลังแก้ไขข้อผิดพลาด:")
display(new_df.head())
display(new_df.info())
print(f'จำนวนแถวใน DataFrame ใหม่: {len(new_df)}')
```

```python
# prompt: อยากทราบว่าคนเสียชีวิตจากสาเหตุอะไรมากที่สุด โดยคิดเป็นค่าเฉลี่ย

# Filter the DataFrame to only include rows where 'dead' is '1' (มีผู้เสียชีวิต)
dead_accidents_df = new_df[new_df['dead'] == '1']

# Group the filtered DataFrame by 'cause' and count the occurrences
dead_cause_counts = dead_accidents_df.groupby('cause').size().reset_index(name='death_count')

# Calculate the total number of deaths across all recorded causes
total_deaths = dead_cause_counts['death_count'].sum()

# Calculate the percentage of deaths for each cause
dead_cause_counts['death_percentage'] = (dead_cause_counts['death_count'] / total_deaths) * 100

# Sort the results in descending order by death_count
most_common_death_causes = dead_cause_counts.sort_values(by='death_count', ascending=False)

print("\nสาเหตุที่ทำให้คนเสียชีวิตมากที่สุด (ค่าเฉลี่ยจากข้อมูลทั้งหมด):")
display(most_common_death_causes)

most_common_death_causes.to_csv('most_common_death_causes.csv', encoding='utf-8-sig', index=False)
```
